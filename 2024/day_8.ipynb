The signal only applies its nefarious effect at specific antinodes based on the resonant frequencies of the antennas. Because the topmost A-frequency antenna overlaps with a 0-frequency antinode, there are 14 total unique locations that contain an antinode within the bounds of the map.

Calculate the impact of the signal. How many unique locations within the bounds of the map contain an antinode?

```python
import numpy as np

def get_cords(char,grid):
    '''Returns coordinates where char appears in grid'''
    
    cords = []

    # itterate through coordinates in grid
    # if item in coordinates equals char append coordinates to list
    for y in range(0,len(grid)):
        for x in range(0,len(grid[0])):
            
            if grid[y][x] == char:
                
                cords.append((y,x))
                
    return cords

# read in text
grid = np.array([list(line.strip()) for line in open('day8ex.txt').readlines()])

# get unique characters in grid
chars = []

for line in grid:
    chars.extend(line)

chars = [char for char in chars if char != '.']

u_chars = list(set(chars))

print("U_chars:", u_chars)

sat_cords = []

# for each unique character
for char in u_chars:

    cords = get_cords(char,grid)
    
    print('start_cords:', cords)
    
    sat_cords.extend(cords)
    # get anchor coordenate
    for r in range(0, len(cords)):
        
        anchor = cords[r]
        
        cut_cords = [cord for cord in cords if cord != anchor]
        
        print('Anchor:', anchor)
        print('Cords:', cut_cords)
    
        # anchor cords
        ya = anchor[0]
            
        xa = anchor[1]
        
        # get reversed path from anchor to each other coordinate
        for cord in cut_cords:
          
            # non-ancher cords
            yb = cord[0]
            
            xb = cord[1]
            
            # path from anchor to non-anchor
            y_diff = ya - yb
            
            x_diff = xa - xb
            
            #reveresd path
            y_diff = -y_diff
            
            x_diff = -x_diff
            
            # reveresd path coordinate
            y = ya + y_diff
                
            x = xa + x_diff
            
            if (y > 0) and (y < len(grid)) and (x > 0) and (x < len(grid[0])):
                
                sat_cords.append((y,x))
                print("Append", (y,x))
                
len(list(set(sat_cords)))
```
